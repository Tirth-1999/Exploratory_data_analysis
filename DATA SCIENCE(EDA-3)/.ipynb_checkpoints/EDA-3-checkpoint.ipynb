## Case Study -2 on Diamond dataset 

```python
dmd_db = pd.read_csv(r"diamonds.csv")
```

```python
from pandas.api.types import CategoricalDtype
from plotnine import *
from plotnine.data import mpg
%matplotlib inline
```

```python
## ploting a histogram for continuously ditributed data  (price)
ggplot(dmd_db, aes(x='price')) + geom_histogram(fill = 'yellow',color = 'black')
```

```python
## plotting a bar-plot for categorical data  like in this case its cut 
ggplot(dmd_db , aes(x= 'cut',fill = 'cut')) + geom_bar()
```

```python
# plotting a scatter-plot for understanding the relationship between 2 numeric coloums (in this case its price with respect to carat)
ggplot(dmd_db,aes(x='carat',y = 'price',color = 'cut'))+geom_point()
```

```python
## plotting a bow-plot to determine how does a numerical value change with repect to categorial value (carat  vs clarity)
ggplot(dmd_db,aes(x='clarity',y = 'carat',fill = 'cut'))+geom_boxplot()
```

```python
## faceting to divide one plot into multiple plot based on one column
ggplot(dmd_db,aes(x='clarity',y = 'carat',fill = 'cut'))+geom_boxplot() + facet_grid('~cut')
```

## 3 EDA- Pokemon dataset


```python
pk_df = pd.read_csv(r"pokemon.csv")
pk_df
```

```python
pk = pk_df.is_legendary.unique()
for i in range(len(pk)):
    temp_pk = pk_df[pk_df.is_legendary == pk[i]]
    print(pk[i],' : ',temp_pk.shape[0])
```

```python
pk = pk_df.generation.unique()
for i in range(len(pk)):
    temp_pk = pk_df[pk_df.generation == pk[i]]
    print(pk[i],' : ',temp_pk.shape[0])
```

```python
pk = pk_df.type1.unique()
for i in range(len(pk)):
    temp_pk = pk_df[pk_df.type1 == pk[i]]
    print(pk[i]," "* (8 - len(pk[i])),' : ',temp_pk.shape[0])
```

```python
##instead of all the above method there is a new method called as .value_counts()
print(pk_df.is_legendary.value_counts())
print("----------****----------")
print(pk_df.type1.value_counts())
print("----------****----------")
print(pk_df.generation.value_counts())
```
